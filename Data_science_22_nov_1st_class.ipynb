estamos buscando la max inertia

```python
original_vector = np.array([[1],[0]])
original_vector
```

```python
angle=np.radians(49.999999999)
#matriz de rotacion
rotation_matrix = np.array([[np.cos(angle), -np.sin(angle)],
                            [np.sin(angle), np.cos(angle)]])
rotation_matrix
```

```python
rotated_vector = np.dot(rotation_matrix, original_vector)
```

```python
plt.subplot(132)
plt.plot(M_r[0,:],M_r[1,:],'ob') # the data
plt.quiver(0, 0, rotated_vector[0, 0], rotated_vector[1, 0], angles='xy', scale_units='xy', scale=1, color='r', label='Vector Rotado')
plt.title("Rotación")
```

```python
#calcular scores
score=np.dot(M_r.T,rotated_vector)

print(score)

# Calcular la varianza de los datos rotados
variance_rotated = np.var(score, axis=0)

print(f'\n  varianza en ejes rotados es:{variance_rotated }')
```
